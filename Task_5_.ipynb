```python
from rdkit import Chem
from rdkit.Chem import Descriptors
import pandas as pd

def lipinski(smiles, verbose=False):
    baseData = []  # List of lists
    i = 0  
    
    for elem in smiles:
        mol = Chem.MolFromSmiles(elem)
        if mol is not None:
            mw = Descriptors.MolWt(mol)
            logp = Descriptors.MolLogP(mol)
            h_donors = Descriptors.NumHDonors(mol)
            h_acceptors = Descriptors.NumHAcceptors(mol)

            baseData.append([mw, logp, h_donors, h_acceptors])  # Append as a list

            if verbose:
                print(f"{i}: Processed {elem} -> MW: {mw}, LogP: {logp}, HDonors: {h_donors}, HAcceptors: {h_acceptors}")
        i += 1

    columnNames = ["MW", "LogP", "NumHDonors", "NumHAcceptors"]
    descriptors = pd.DataFrame(data=baseData, columns=columnNames)  # Ensure 2D structure
    
    return descriptors
```

```python
import numpy as np
from rdkit import Chem

df = df.copy()

# Convert literal "nan" strings back to real NaN
df["canonical_smiles"] = df["canonical_smiles"].replace("nan", np.nan)

# Drop missing SMILES
df = df.dropna(subset=["canonical_smiles"])

# Strip whitespace
df["canonical_smiles"] = df["canonical_smiles"].astype(str).str.strip()

# RDKit validation
df["mol"] = df["canonical_smiles"].apply(Chem.MolFromSmiles)

# Drop invalid molecules
df = df[df["mol"].notna()]

# Remove helper column
df = df.drop(columns="mol")

print("Final molecules:", len(df))
```

```python
import numpy as np

df["standard_value"] = pd.to_numeric(df["standard_value"], errors="coerce")
df = df.dropna(subset=["standard_value"])
df["pIC50"] = -np.log10(df["standard_value"] * 1e-6)
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski
from rdkit.Chem import rdMolDescriptors


df["mol"] = df["canonical_smiles"].apply(Chem.MolFromSmiles)
df = df[df["mol"].notna()]

```

```python
df["MW"] = df["mol"].apply(Descriptors.MolWt)
df["LogP"] = df["mol"].apply(Descriptors.MolLogP)
df["HBD"] = df["mol"].apply(Descriptors.NumHDonors)
df["HBA"] = df["mol"].apply(Descriptors.NumHAcceptors)

df["TPSA"] = df["mol"].apply(rdMolDescriptors.CalcTPSA)
df["RotBonds"] = df["mol"].apply(Lipinski.NumRotatableBonds)
```

```python
df["Lipinski_pass"] = (
    (df["MW"] <= 500) &
    (df["LogP"] <= 5) &
    (df["HBD"] <= 5) &
    (df["HBA"] <= 10)
)

df["Veber_pass"] = (
    (df["TPSA"] <= 140) &
    (df["RotBonds"] <= 10)
)

df_filtered = df[df["Lipinski_pass"] & df["Veber_pass"]]
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import TanimotoSimilarity

# Reference: Ciprofloxacin SMILES
cipro_smiles = 'C1CC1N2C=C(C(=O)C3=CC(=C(C=C32)N4CCN(CC4)F)F)C(=O)O'
cipro = Chem.MolFromSmiles(cipro_smiles)
if cipro is None:
    raise ValueError("Invalid reference SMILES for ciprofloxacin")

cipro_fp = AllChem.GetMorganFingerprintAsBitVect(cipro, radius=2, nBits=2048)

def tanimoto_to_cipro(mol):
    if mol is None:
        return 0.0
    try:
        fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
        return TanimotoSimilarity(cipro_fp, fp)
    except Exception as e:
        print(f"Error computing fingerprint: {e}")
        return 0.0

# Make sure ROMol column exists (should already be there from previous step)
if 'ROMol' not in final_hits.columns:
    final_hits['ROMol'] = final_hits['canonical_smiles'].apply(Chem.MolFromSmiles)
    final_hits = final_hits.dropna(subset=['ROMol'])

# Apply similarity
final_hits['Tanimoto_to_cipro'] = final_hits['ROMol'].apply(tanimoto_to_cipro)

# Sort: lowest similarity = most novel
novel_hits = final_hits.sort_values('Tanimoto_to_cipro', ascending=True)

# Display top 10 most novel
print("Top 10 most novel hits (lowest Tanimoto similarity to ciprofloxacin):")
display(novel_hits[[
    'molecule_chembl_id', 
    'canonical_smiles', 
    'pIC50', 
    'Tanimoto_to_cipro', 
    'MW', 'LogP', 'TPSA', 
    'Rationale'
]].head(10).round(3))

# Optional: save
novel_hits.to_csv("Task5_shortlisted_hits_with_tanimoto.csv", index=False)
```
